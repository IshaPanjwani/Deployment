```python
#converting datatype of dates

cab_data['Date of Travel'] = pd.to_datetime(cab_data['Date of Travel'])
```

```python
#seperating months and years

months = []
years = []
for i in range(len(cab_data['Date of Travel'])):
    months.append(cab_data['Date of Travel'][i].month)
    years.append(cab_data['Date of Travel'][i].year)
cab_data['Month'] = months
cab_data['Year'] = years
```

```python
cab_data
```

```python
cab_data.info()
```

```python
#consolidating into master dataframe

master_data = cab_data.merge(transactionID_data, on = 'Transaction ID').merge(customer_ID, on = 'Customer ID').merge(city_data, on = 'City')
```

```python
master_data
```

```python
master_data.info()
```

```python
master_data.isnull().sum()
```

```python
for column in master_data.columns:
    if ' ' in column:
        master_data = master_data.rename(columns = {column:column.replace(' ', '_')})
        
for column in ['Population','Users']:
    master_data[column] = master_data[column].str.replace(',', '')    
```

```python
master_data
```

```python
for column in ["Company", "City", "Payment_Mode", "Gender"]:
    master_data[column] = master_data[column].astype('category')

for column in ["Population", "Users"]:
    master_data[column] = master_data[column].astype('int64')
```
